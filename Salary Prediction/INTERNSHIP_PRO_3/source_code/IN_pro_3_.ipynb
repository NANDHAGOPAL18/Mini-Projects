```python
import pandas as pd 
df=pd.read_csv('/content/hiring_pro_3')
df
```

```python
import math
test_median=math.floor(df['test_score(out of 10)'].median())
test_median
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(test_median)
df
```

```python
df['experience']=df['experience'].fillna('zero')
df
```

```python
from word2number import w2n
```

```python
w2n.word_to_num('two million three thousand nine hundred and eighty four')
```

```python
#approach_1
def convertword2number(x):
  return w2n.word_to_num(x)
```

```python
df['experience']=df['experience'].apply(convertword2number)
df
```

```python
x=df.drop('salary($)',axis=1)
x
```

```python
y=df['salary($)']
y
```

```python
#scikit learn
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[3,5,6]])
```

#understanding the ML

```python
reg.coef_
```

```python
b=reg.intercept_
b
```

```python
y=(m1*3)+(m2*5)+(m3*6)+b
y
```
