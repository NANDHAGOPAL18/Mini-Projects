```python
def common():
    df = pd.read_csv('price_finding_dataset.csv')
    X = df.drop('price',axis=1)
    y = df['price']
    reg = linear_model.LinearRegression()
    reg.fit(X,y)
    reg.predict([[1,8,512,6000]])
    m = reg.coef_
    m1 = -8.15615987e+02
    m2 = 2.91635764e+03
    m3 = 2.18440977e+01
    m4 = -3.54758962e-01
    b = reg.intercept_
    Y = (-8.15615987e+02*1)+(2.91635764e+03*8)+(2.18440977e+01*512)+(-3.54758962e-01*6000)+(-1244.5405850845382)
    ytest = reg.predict(X)
    age = int(input("enter the mobile age :"))
    ram = int(input("enter the mobile Ram :"))
    rom = int(input("enter the mobile Rom :"))
    battery = int(input("enter the mobile battery :"))
    final_price = reg.predict([[age,ram,rom,battery]])
    print("the predicted price of your mobile :",final_price)
    
    
    p1 = int(input("how many products you want :"))
    if (p1>2 and p1<=10):
        d2 = final_price * 10/100
        print ("you got 10% offer :",final_price - d2)
    elif (p1>10 and p1<=30):
        d3 = final_price * 25/100
        print ("you got 25% offer :",final_price - d3)
    elif (p1>30):
        d4 = final_price * 35/100
        print ("you got 35% offer :",final_price - d4)
    else :
        d1 = final_price * 5/100
        print("you got 5% offer : " , final_price - d1)
```
