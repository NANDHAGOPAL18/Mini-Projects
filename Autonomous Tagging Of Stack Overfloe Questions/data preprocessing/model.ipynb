```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/NANDY/Downloads/Naan Mudhalvan/datasets/final-data.csv')
df.head()
```

```python
import ast 
df['Tags'] = df['Tags'].apply(lambda x : ast.literal_eval(x))
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split

from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
```

```python
y = df['Tags']
y
```

```python
multilabel = MultiLabelBinarizer()
y = multilabel.fit_transform(y)
```

```python
multilabel.classes_
```

```python
pd.DataFrame(y,columns = multilabel.classes_)
```

```python
tfidf = TfidfVectorizer(analyzer = 'word' , max_features = 1000 , ngram_range=(1,3),stop_words='english')
X = tfidf.fit_transform(df['Body'].values.astype(str))
X.shape , y.shape
```

```python
X_train , X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
tfidf.vocabulary_
```

```python
sgd = SGDClassifier()
lr = LogisticRegression()
svc = LinearSVC()
```

```python
def j_score(y_true,y_pred):
    jaccard = np.minimum(y_true,y_pred).sum(axis=1)/np.maximum(y_true,y_pred).sum(axis=1)
    return jaccard.mean()*100

def print_score(y_pred,clf):
    print("CLF: ",clf.__class__.__name__)
    print("Jaccard score: {}".format(j_score(y_test,y_pred)))
    print("=================================================")
```

```python
for classifier in[sgd,lr,svc]:
    clf = OneVsRestClassifier(classifier)
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    print_score(y_pred,classifier)
```

```python
import joblib
joblib_file = "tagPredictorr.pkl"
joblib.dump(clf,joblib_file)
```

```python
tagPredictorModel = joblib.load("tagpredictorr.pkl")
```
